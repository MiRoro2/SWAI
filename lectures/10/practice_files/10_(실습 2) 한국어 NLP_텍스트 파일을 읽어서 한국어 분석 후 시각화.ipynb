3. 문장에서 품사 및 명사 추출하기 (치환  or 제거 명사 포함)

<span style="color:blue">  
● pos()함수는 문장에서 품사 정보를 출력하는 함수이다. <br>
   pos()는 okt, kkma 모든 모듈에서 사용가능하다.<br>
   pos()함수의 입력은 string 타입, 리턴은 list 타입이다.<br>
    

     
okt.pos()의 경우, 리턴 값은  <br>
Noun : 명사, Josa : 조사, Punctuation : 구두점, Adjective : 형용사, Verb : 동사 등등.. 

kkma.pos()의 경우, 리턴 값은  <br> 
NP : 대명사, JX : 보조사, NNG : 보통명사, NNP: 고유명사, SP : 쉼표,가운뎃점,콜론,빗금, <br> 
JKS : 주격조사, VA : 형용사, VV : 동사, EFN : 평서형 종결 어미 등등..     
</span> 





<span style="color:blue">    
● nouns()함수는 입력된 문장에서 명사를 출력하는 함수이다.  <br>     
   nouns()는 okt, kkma 모든 모듈에서 사용가능하다. <br>  
   nouns()함수의 입력은 string 타입, 리턴은 list 타입이다. <br>     
</span> 

<span style="color:blue">  
okt.nouns() 경우 : 문장에서 중복된 명사를 모두 출력한다. <br> 
kkma.nouns()의 경우 : 문장에서 중복된 명사는 1번만 출력한다.
</span>

<span style="color:blue; font-size: 14pt"> 
 3.1. okt.pos() 품사 출력하기 / okt.nouns() 명사 출력하기 <b> (input type : string, return type : list) 
</span>

```python
okt_nouns=[]
line=0

for i in data_readlines :  
        print(line,"line", type(i))     
        print("문장 출력 : ", i)
        
        #품사 출력 부분
        print("품사 출력 : ", end='') 
        print(okt.pos(i), "\n") #문장의 품사 출력, pos(i)에서 i는 string타입의 문장 데이터
        
        #명사 출력 부분
        data = okt.nouns(i)  #okt.nouns()를 활용하여 명사를 모두 추출하여 data 리스트에 저장 
        print("문장에서 명사 출력 : ", end='') 
        print(data, "\n")    
  
        for j in range(0,len(data)) : 
            print("명사", [j], ": ", end='')
            print(data[j])
            
            #치환 명사가 있다면 (가령 "젤"을 '제일'로 치환한다면)
            if data[j] == '젤' :  
                count = 0
                data_j = okt.nouns(data[j].replace('젤','제일'))  #okt.nouns()를 활용하여 '젤'명사를 '제일'명사로 치환
                print('●', data[j], "→" , data_j, "로 치환됨")  
                okt_nouns.append(data_j[count])
                count += 1
            else :    
                okt_nouns.append(data[j])
            
            
            print("치환 명사를 고려한 최종 명사 리스트 : ", end='' )           
            print(okt_nouns, "\n")
            
        line += 1  
        print("치환 명사를 고려한 최종 명사 리스트 길이 : ",len(okt_nouns)) 
        print('-'*100, '\n')
```

 ● okt의 경우 아래와 같은 명사 리스트로 최종 완성

```python
print(okt_nouns) #okt_nouns 치환 명사를 고려한 최종 명사 리스트 출력  
```

```python
print(data_readlines)  #위에서 readlines()로 읽은 후 저장해 놓은 data_readlines는 리스트 타입임을 확인
print(type(data_readlines))
```

```python
#kkma.nouns()도 string 타입을 사용함
#data_readlines는 리스트 타입인 상황

kkma_nouns =[]
line= 0

for i in data_readlines :
        print(line,"line", type(i))     
        print("문장 출력 : ", i)
        
        #품사 출력 부분
        print("품사 출력 : ", end='') 
        print(kkma.pos(i), "\n") #문장의 품사 출력, pos(i)에서 i는 string타입의 문장 데이터
        
        #명사 출력 부분
        data = kkma.nouns(i)  #kkma.nouns()를 활용하여 명사를 모두 추출하여 data 리스트에 저장 
        print("문장에서 명사 출력 : ", end='') 
        print(data, "\n")       
        
        for j in range(0,len(data)) : 
            print("명사", [j], ": ", end='')
            print(data[j])                       
            
            #치환 명사가 있다면 (가령 "젤"을 '제일'로 치환한다면)
            if data[j] == '짱' :                 
                data_j = kkma.nouns(data[j].replace('짱','제일'))  #kkma.nouns()를 활용하여 '젤'명사를 '제일'명사로 치환
                print('●', data[j], "→" , data_j, "로 치환됨")  
                kkma_nouns.append(data_j[0])
            else :    
                kkma_nouns.append(data[j])            
            
            print("치환 명사를 고려한 최종 명사 리스트 : ", end='' )           
            print(kkma_nouns, "\n")        
 
        line += 1 
        print("리스트 길이 : ",len(kkma_nouns)) 
        print('-'*100)
        
#4라인 '짱'을 'NNG(보통명사)'로 해석하여 명사로 추출됨
#5라인'블루베리'가 ('블루', 'NNG'), ('베리', 'NNP')으로 파싱된 후, '블루', '블루베리', '베리'가 명사로 추출됨        
```

<span style="color: blue; font-size: 14pt">
6.1. 불용어 사전을 통과한 okt.nouns()를 대상으로 글자 수 제한으로 불용어 제거하기

```python
print("● okt.nouns()를 활용하여 추출된 명사 중에서 불용어 사전을 통과한 결과 : %d 개" %len(non_stop_okt))
print(non_stop_okt)
```

```python
refine_okt = []
for i in non_stop_okt :  #불용어에 해당하지 않는 단어를 대상으로 
    if len(i) >= 2 and len(i) <= 10 :  #글자수가 2자 ~ 10자에 해당되는 단어만 추출하여 final_okt에 추가한다.
        refine_okt.append(i) 
print("● 추출된 명사 중에서 글자 수가 2~10 사이에 해당하는 명사 : %d 개" %len(refine_okt))    
print(refine_okt)

#'짱' 단어가 글자수가 1개 이므로 추출된 명사에서 삭제됨
```

6.2. 불용어 사전을 통과한 kkma.nouns()를 대상으로 글자 수 제한으로 불용어 제거하기

```python
print("● kkma.nouns()를 활용하여 추출된 명사 중에서 불용어 사전을 통과한 결과 : %d 개" %len(non_stop_kkma))
print(non_stop_kkma)
```

```python
refine_kkma = []
for i in non_stop_kkma :  #불용어에 해당하지 않는 단어를 대상으로     
    if len(i) >= 2 and len(i) <= 10 :  #글자수가 2자 ~ 10자에 해당되는 단어만 추출하여 final_kkma에 추가한다.
        refine_kkma.append(i) 
print("● 추출된 명사 중에서 글자 수가 2~10 사이에 해당하는 명사 : %d 개" %len(refine_kkma))    
print(refine_kkma)

#'짱' 단어가 글자수가 1개 이므로 추출된 명사에서 삭제됨
```

```python
print("okt.nouns()의 경우")

final_okt = Counter(refine_okt)
print("● 불용어가 제거된 명사의 빈도수 출력")
print(type(final_okt))
print(final_okt)

final_okt_sort = final_okt.most_common(10) #most_common() 내림차순으로 정렬한다.()안의 숫자는 상위 몇개를 추출할것인지에 해당
print("● 내림차순 정렬")
print(type(final_okt_sort))
print(final_okt_sort)

final_okt_sort_dic = dict(final_okt_sort)
print("● 딕셔너리 타입으로 변환 ")
print(final_okt_sort_dic)
print(type(final_okt_sort_dic))
```
