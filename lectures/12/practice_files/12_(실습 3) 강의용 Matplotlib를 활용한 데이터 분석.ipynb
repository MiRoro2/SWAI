# Matplotlib.pyplot를 활용한 데이터 분석

```python
import matplotlib.pyplot as plt

#plot() 함수는 리스트의 값들이 y 값들이라고 가정하고, x 값 [0, 1, 2, 3]을 자동으로 생성 
plt.plot([1, 2, 3, 4])
plt.show() 
```

```python
# x축과 y축을 기반으로 그래프
plt.plot([1, 2, 3, 4], [1, 4, 9, 16])
plt.show()
```

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'ro') #red 색상으로'o' 원형 
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'b-') #blue 색상으로'-' 원형 
#plt.axis([0, 10, 0, 20])
plt.show()
```

```python
plt.figure(figsize=(10,3))  #figure() 호출할 때 마다 그래프를 그릴 수 있는 사각형 프레임이 생성


#figure 메소드로 figure을 작성하고, subplot 메소드로 axes를 생성

plt.subplot(1,3,1)  #figure()로 인해 생성된 사각형 프레임 내에서 1행 3열 중에서 첫번째 프레임을 의미
plt.subplot(1,3,2)  #figure()로 인해 생성된 사각형 프레임 내에서 1행 3열 중에서 두번째 프레임을 의미
plt.subplot(1,3,3)  #figure()로 인해 생성된 사각형 프레임 내에서 1행 3열 중에서 세번째 프레임을 의미
plt.show()
```

# 시각화를 위해 numpy를 이용하여 데이터셋 생성 

```python
vegetables = ["cucumber", "tomato", "lettuce", "asparagus", "potato", "wheat", "barley"]
#vegetables = [ "오이", "토마토", "상추", "아스파라거스", "감자", "밀", "보리"]

print(vegetables)
print(type(vegetables))
print(len(vegetables),"종류의 채소 데이터")
```

```python
farmers = ["Farmer Joe", "Upland Bros", "Smith Gardening", 
           "Agrifun", "Organiculture", "BioGoods Ltd", "Cornylee Corp."]

print(farmers)
print(type(farmers))
print(len(farmers),"명의 농부 데이터")
```

```python
harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [0.1, 2.0, 0.0, 1.4, 0.0, 1.9, 6.3]])

print(harvest)
print(type(harvest))
print(len(harvest),"라인", len(harvest[0]),"컬럼의 데이터")
```

# 시각화 

```python
np.arange(7)
```

```python
#setp 1 -----------------------------------------------------------------
fig, ax = plt.subplots(figsize=(10, 6)) 
print(fig)  #프레임 생성
print(ax)   #1개의 축 생성  

#setp 2 -----------------------------------------------------------------
ax.set_title("Harvest of local farmers (in tons/year)")
ax.imshow(harvest, cmap='ocean')  #numpy.ndarray 타입 사용 가능 
#ax.imshow(harvest)  #numpy.ndarray 타입 사용 가능 

#setp 3 -----------------------------------------------------------------

#xaxis, yaxis의 눈금 위치를 설정 
print('● len(farmers)은', len(farmers), '이고, 값은  ', end='')
print(np.arange(len(farmers)))

print('● len(vegetables)은', len(vegetables),  '이고, 값은  ',end='')
print(np.arange(len(vegetables)))

ax.set_xticks(np.arange(len(farmers)))      #x축 [0 1 2 3 4 5 6]
ax.set_yticks(np.arange(len(vegetables)))   #y축 [0 1 2 3 4 5 6]

#setp 4 -----------------------------------------------------------------

#각각의 x, y축에 레이블링
ax.set_xticklabels(farmers)     #x축 [0 1 2 3 4 5 6]부분에 레이블링이 됨
ax.set_yticklabels(vegetables)  #y축 [0 1 2 3 4 5 6]부분에 레이블링이 됨

#setp 5 -----------------------------------------------------------------

#x축 레이블을 회전하고 정렬 
#ha = 'right'는 수평 레이블 텍스트의 오른쪽 끝을 눈금에 맞춥니다.
#ha = 'left'는 수평 레이블 텍스트의 왼쪽 끝을 눈금에 맞춥니다.
#ha = 'center'는 수평 레이블 텍스트의 중심을 눈금에 맞춥니다.

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", fontsize=12)
#plt.setp(ax.get_xticklabels(), rotation=45, ha="left", fontsize=12)
#plt.setp(ax.get_xticklabels(), rotation=45, ha="center", fontsize=12)

#y축 레이블을 정렬 
plt.setp(ax.get_yticklabels(), ha="right", fontsize=12)

#setp 6 -----------------------------------------------------------------
#ha 수평 축
#va 수직 축 

for i in range(len(vegetables)):  #column
    for j in range(len(farmers)):   #row
        text = ax.text(j, i, harvest[i, j], ha="center", va="center", color="y", fontsize=14)
        print(text, end='')
        print('-->', harvest[i, j])
    print('-'*50)               
        
#-------------------------------------------------------------------------        
plt.show()
```

# 한글을 포함한 그래프 그리기 

```python
#한글 사용을 위해서 matplotlib.font_manager 사용

import matplotlib
import matplotlib.font_manager as fm 


font_location="c:/windows/fonts/HMFMMUEX.TTC"  #한글 폰트를 셋팅함

font_name=fm.FontProperties(fname=font_location).get_name() 

matplotlib.rc('font',family=font_name) #한글 깨짐을 방지하기 위해서 반드시 해야할 부분
```

```python
#vegetables = ["cucumber", "tomato", "lettuce", "asparagus", "potato", "wheat", "barley"]
vegetables = [ "오이", "토마토", "상추", "아스파라거스", "감자", "밀", "보리"]

farmers = ["Farmer Joe", "Upland Bros.", "Smith Gardening", "Agrifun", "Organiculture", "BioGoods Ltd.", "Cornylee Corp."]

harvest = np.array([[0.8, 2.4, 2.5, 3.9, 0.0, 4.0, 0.0],
                    [2.4, 0.0, 4.0, 1.0, 2.7, 0.0, 0.0],
                    [1.1, 2.4, 0.8, 4.3, 1.9, 4.4, 0.0],
                    [0.6, 0.0, 0.3, 0.0, 3.1, 0.0, 0.0],
                    [0.7, 1.7, 0.6, 2.6, 2.2, 6.2, 0.0],
                    [1.3, 1.2, 0.0, 0.0, 0.0, 3.2, 5.1],
                    [6.3, 2.0, 0.0, 1.4, 0.0, 1.9, 0.1]])
```

```python
#setp 1 -----------------------------------------------------------------
fig, ax = plt.subplots(figsize=(10, 6)) 
print(fig)  #프레임 생성
print(ax)   #1개의 축 생성  

#setp 2 -----------------------------------------------------------------
ax.set_title("지역 농민의 수확 (톤 / 년)", fontsize=16)
#ax.imshow(harvest, cmap='ocean')  #numpy.ndarray 타입 사용 가능 
ax.imshow(harvest)  #numpy.ndarray 타입 사용 가능 

#setp 3 -----------------------------------------------------------------

#xaxis, yaxis의 눈금 위치를 설정 
print('● len(farmers)은', len(farmers), '이고, 값은  ', end='')
print(np.arange(len(farmers)))

print('● len(vegetables)은', len(vegetables),  '이고, 값은  ',end='')
print(np.arange(len(vegetables)))

ax.set_xticks(np.arange(len(farmers)))      #x축 [0 1 2 3 4 5 6]
ax.set_yticks(np.arange(len(vegetables)))   #y축 [0 1 2 3 4 5 6]


#setp 4 -----------------------------------------------------------------

#각각의 x, y축에 레이블링
ax.set_xticklabels(farmers)     #x축 [0 1 2 3 4 5 6]부분에 레이블링이 됨
ax.set_yticklabels(vegetables)  #y축 [0 1 2 3 4 5 6]부분에 레이블링이 됨

#setp 5 -----------------------------------------------------------------

#x축 레이블을 회전하고 정렬 
#ha = 'right'는 수평 레이블 텍스트의 오른쪽 끝을 눈금에 맞춥니다.
#ha = 'left'는 수평 레이블 텍스트의 왼쪽 끝을 눈금에 맞춥니다.
#ha = 'center'는 수평 레이블 텍스트의 중심을 눈금에 맞춥니다.

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", fontsize=12)
#plt.setp(ax.get_xticklabels(), rotation=45, ha="left", fontsize=12)
#plt.setp(ax.get_xticklabels(), rotation=45, ha="center", fontsize=12)

#y축 레이블을 정렬 
plt.setp(ax.get_yticklabels(), ha="right", fontsize=12)

#setp 6 -----------------------------------------------------------------
#ha 수평 축
#va 수직 축 

for i in range(len(vegetables)):  #column
    for j in range(len(farmers)):   #row
        text = ax.text(j, i, harvest[i, j], ha="center", va="center", color="w", fontsize=14)
        print(text, end='')
        print('-->', harvest[i, j])
    print('-'*50)               
        
#-------------------------------------------------------------------------        
plt.show()
```
