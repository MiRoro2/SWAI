```python
import matplotlib.pyplot as plt 

plt.rcParams["figure.figsize"] = (14,4) #가로 너비 14인치, 세로 높이 4인치 

#%matplotlib qt5

#pandas.core.frame.DataFrame 타입도 plot()를 이용한 그래프가 출력된다.
#각 열(column) 단위에 해당하는 년도별로 South SUM, North SUM 그래프가 출력 
df_sn.plot()  
```

```python
# 행, 열 전치하여 다시 그리기
# 출력된 그래프를 시간의 흐름에 따른 연도별 발전량 변화 추이를 보기 위해서는 
# 연도 값을 x축에 표시하는 것이 좋습니다.  
# 연도값을 x 축으로 바꾸기위해서 열레이블을 구성하고 있는 년도에 해당하는 값이 
# 행레이블에 위치하도록 행렬을 전치합니다. 

tdf_sn = df_sn.T  # T는 인덱스와 컬럼을 바꾼다. 
tdf_sn
```

```python
tdf_sn.info()
```

```python
tdf_sn.plot()    # plot 그래프
plt.show()       
```

```python
tdf_sn.plot(kind='bar')   # 막대 그래프
plt.show()   
```

```python
tdf_sn.info()
```

```python
tdf_sn
```

```python
tdf_sn.plot(kind='hist')
```

```python
tdf_sn.plot(kind='hist', bins=100, color=('coral','#345633'))   #히스토그램
plt.show()   

# bins는 x축 구간의 개수를 의미하며 숫자가 커질수록 더 세밀하게 관찰할 수 있습니다. 
# bins=1000 은 x축을 1000개로 쪼개서 분석했다는 뜻입니다. 
```

<span style="color:blue">  
South SUM과 North SUM의 산점도(Scatter plot) 그래프
</span>  

```python
# pandas로 부터 그래프 출력
tdf_sn.plot('South SUM', 'North SUM',  # x축, y축을 의미함 -> 범례에 y축 레이블이 출력됨
         linestyle='none',     # linestyle='-', (실습 권장)
         marker='.',           # marker='x', (실습 권장)
         markersize=10,        # markersize=1, (실습 권장)
         color='blue',         # color='red', (실습 권장)
         alpha=1)              # alpha는 투명도를 의미함,  alpha=0.5 또는 alpha=0.1 (실습할 것)

plt.show()   
```

```python
#pandas로 부터 그래프 출력

plt.rcParams["figure.figsize"] = (9,5) #가로 너비 9인치, 세로 높이 5인치 

tdf_sn.plot('South SUM', 'North SUM',  # x축, y축을 의미함 -> 범례에 y축 레이블이 출력됨
         linestyle='none',        # linestyle='-', (실습 권장)
         marker='.',              # marker='x', (실습 권장)
         markersize=10,           # markersize=1, (실습 권장)
         color='#ff56ff',         # color='red', (실습 권장)
         alpha=1)                 # alpha는 투명도를 의미함,  alpha=0.5 또는 alpha=0.1 (실습할 것)

# pandas로 부터 출력된 그래프를 수정 -> import matplotlib.pyplot as plt 으로부터 plt를 사용
plt.title('남북한 발전 전력량', fontsize=20)
plt.xlabel('South value', fontsize=14)
plt.ylabel('North value', fontsize=14)
plt.show()
```

```python
# pandas로 부터 그래프 출력
tdf_sn.plot('North SUM', 'South SUM',   # x축, y축을 의미함 -> 범례에 y축 레이블이 출력됨
         linestyle='none', 
         marker='o', 
         markersize=30,
         color='blue', 
         alpha=0.5)

# pandas로 부터 출력된 그래프를 수정 -> import matplotlib.pyplot as plt 으로부터 plt를 사용
plt.title('남북한 발전 전력량', fontsize=20)
plt.xlabel('북한 발전 전력량',  fontsize=14)
plt.ylabel('남한 발전 전력량',  fontsize=14)
plt.show()
```
