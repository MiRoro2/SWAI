### x축을 차중, y축을 연비로 하여 구성한 UCI 자동차 연비 데이터셋의 산점도 그래프

```python
# setup
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager 

file_path = "./auto-mpg.xlsx"
data = pd.read_excel(file_path)

font_path = "./fonts/Pretendard-Regular.otf"
font = font_manager.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font.get_name()
plt.rcParams['font.size'] = 12
```

```python
# indexing dataset
data.columns = ['MPG', 'Cylinders', 'Displacement', 'Horsepower', 'Weight', 'Acceleration', 'Year', 'Origin', 'Model']
```

```python
# plot the scatter plot
plt.figure(figsize=(10, 6))
scatter_plot = plt.scatter(data['Weight'], data['MPG'], c="blue")
plt.title('자동차 연비')
plt.xlabel('차중')
plt.ylabel('연비')
plt.grid(True)
plt.show()
```
